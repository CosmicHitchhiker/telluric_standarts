```python
t = ["\n \
format tid\n \
query ","\n \
result count\n \
query sample (region(circle,",",25s))\n \
query sample (region(circle,",",1.5m) & otype='Star' & Vmag<",")\n"]

df['Script'] = t[0] + df['name'] + t[1] + df['name'] + t[2] + df['name'] + t[3] + df['Vmag_lim'] + t[4]

for i in df['Script']:
    print(i)
```

```python
with open("simbad_scr2.txt", 'w', encoding='ascii') as scr:
    scr.write('format object tid "\\n|%OBJECT"\n\n#query begins"')
    for i in df['Script']:
        scr.write(i)
    scr.write('\n')
    scr.write('format display')
```
